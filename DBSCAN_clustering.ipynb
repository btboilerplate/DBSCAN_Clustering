```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
x,y = make_moons(n_samples=250,noise=0.08)
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
# Standard Scaling
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
x_scalar = scalar.fit_transform(x)
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.3,min_samples=5)
x_labels = dbscan.fit_predict(x_scalar)
```

```python
# plt.scatter(x_scalar[:,0],x_scalar[:,1],c=dbscan.labels_)
plt.scatter(x_scalar[:,0],x_scalar[:,1],c=x_labels)
```

```python
from sklearn.metrics import silhouette_score
mask = x_labels != -1
silhouette_score(x_scalar[mask], x_labels[mask])
```

```python
from sklearn.datasets import make_circles
x2, y2 = make_circles(n_samples=300,factor=0.5,noise=0.05,random_state=42)
plt.scatter(x2[:,0],x2[:,1])
```

```python
# Standard Scaling
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
x_scalar2 = scalar.fit_transform(x2)

from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.4,min_samples=3)
x_labels2 = dbscan.fit_predict(x_scalar2)
```

```python
plt.scatter(x_scalar2[:,0],x_scalar2[:,1],c=x_labels2)
```

```python
from sklearn.metrics import silhouette_score
mask2 = x_labels2 != -1
silhouette_score(x_scalar2[mask2], x_labels2[mask2])
```
